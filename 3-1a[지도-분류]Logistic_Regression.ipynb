## X, y 데이터로 분할

```python
# X = df.loc[행, 열]
# X = df.loc[:, ['Gender', 'Age', 'EstimatedSalary']]
X = df.loc[:, 'Gender' : 'EstimatedSalary']
X.head()
```

```python
# y
y = df['Purchased']
y
```

```python
X['Gender'].value_counts()
```

```python
# map() 함수 사용
mapping = {'Male': 0, 'Female':1}
X['Gender'] = X['Gender'].map(mapping)
X.head()
```

## Feature scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# 객체 생성
X_scaler = StandardScaler()
X_scaler
```

```python
# X_scaler.fit(X)
# X_scaler.transform(X)
# 위의 두 문장을 한꺼번에 실행하는 매서드
X_scaled = X_scaler.fit_transform(X)
X_scaled
```

# 학습, 테스트 데이터 분할

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, 
    test_size=0.25, 
    random_state=2025,
    stratify=y # test_size=0.25 분할 비율을 y를 기준으로 분류함
)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

# 모델 학습 및 평가

## 모델 객체 생성

```python
from sklearn.linear_model import LogisticRegression
```

```python
# 모델 객체 생성 (빈 깡통 ai)
model_logrc = LogisticRegression()
model_logrc
```

```python
# 빈깡통 ai여서 없음, 정상오류
model_logrc.coef_
```

## 학습하기

```python
model_logrc.fit(X_train, y_train)
```

```python
# w1, w2, w3 (weight)
model_logrc.coef_
```

```python
# w0 (bias)
model_logrc.intercept_
```

```python
y_prob = model_logrc.predict_proba(X_test)
```

## 모델 테스트

```python
# 모의고사 오는 것
y_pred = model_logrc.predict(X_test)
y_pred
```

```python
# y_prob , y_prob df 로 비교하기
y_result = pd.DataFrame(y_prob)
y_result['y_pred'] = y_pred
```

```python
#  확률값이 0.5 이상이면 0, 0.5 이하면 1로 분류된다.
y_result.tail(20)
```

# 모델 평가하기

- 지도학습은 기본적으로 정확도를 확인함
### 정확도(accuracy)

```python
# 정확도 계산
(y_test == y_pred).sum() /len(y_test)*100
```

### 정밀도 (precision), 재현율(적중율, recall)

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import recall_score, precision_score
from sklearn.metrics import classification_report
```

```python
# 재현율
recall_score(y_test, y_pred)
```

```python
#  정밀도
precision_score(y_test, y_pred)
```

```python
# ml의 분류제 종합 평가 점수
print(classification_report(y_test, y_pred))
```

# 분류 결과 시각화 하기
- Gender를 고정값으로 설정

```python
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap

# 3D → 2D 차원 축소
pca = PCA(n_components=2)
X_test_2d = pca.fit_transform(X_test)

# 2D meshgrid
X1, X2 = np.meshgrid(
    np.arange(X_test_2d[:, 0].min() - 1, X_test_2d[:, 0].max() + 1, 0.01),
    np.arange(X_test_2d[:, 1].min() - 1, X_test_2d[:, 1].max() + 1, 0.01)
)

# 원래 3D 공간으로 역변환 후 예측
X_grid_2d = np.c_[X1.ravel(), X2.ravel()]
X_grid_3d = pca.inverse_transform(X_grid_2d)
predictions = model_logrc.predict(X_grid_3d).reshape(X1.shape)

# 시각화
plt.figure(figsize=(10, 7))
plt.contourf(X1, X2, predictions,
             alpha=0.75, cmap=ListedColormap(('red', 'green')))

for i, j in enumerate([0, 1]):
    plt.scatter(X_test_2d[y_test == j, 0], 
               X_test_2d[y_test == j, 1],
               c='darkred' if j == 0 else 'darkgreen',
               label=f'Purchased={j}',
               edgecolors='white',
               s=100,
               linewidth=1.5)

plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})', fontsize=12)
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})', fontsize=12)
plt.title('Decision Boundary (PCA)', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# 1) 모델에서 예측 확률 얻기 (양성 클래스 확률)
y_prob = model_logrc.predict_proba(X_test)[:, 1]

# 2) ROC Curve 계산
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# 3) AUC 계산
auc_score = roc_auc_score(y_test, y_prob)
print("AUC:", auc_score)

# 4) ROC Curve 시각화
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc_score:.3f})")
plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()
```
